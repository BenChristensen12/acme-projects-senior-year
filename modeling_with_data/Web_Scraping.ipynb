Parse through html code to get data

```python
import requests
import re
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import os
from heapq import nlargest
```

```python
def prob1():
    """
    Use the requests library to get the HTML source for the website http://www.example.com.
    Save the source as a file called 'example.html'. If the file already exists, make
    sure not to scrape the website, or overwrite the file.
    """
    filename = "example.html"
    #Check if the file already exists to prevent scraping unnecessarily
    file_exists = os.path.isfile(filename)
    if file_exists:
        print("Did nothing because the file already exists!")
    else:
        #Scrape the website for HTML
        response = requests.get("http://www.example.com")
        #Write the HTML contents to the file
        with open(filename, 'w') as outfile:
            outfile.write(response.text)


# Problem 2
def prob2():
    """Examine the source code of http://www.example.com. Determine the names
    of the tags in the code and the value of the 'type' attribute associated
    with the 'style' tag.

    Returns:
        (set): A set of strings, each of which is the name of a tag.
        (str): The value of the 'type' attribute in the 'style' tag.
    """
    return set(["html", "head", "title", "meta", "style", "body", "div", "h1", "p", "a"]), "text/css"

# Problem 3
def prob3(code):
    """Return a list of the names of the tags in the given HTML code."""
    tags = BeautifulSoup(code, "html.parser").find_all(True)
    return [tag.name for tag in tags]


# Problem 4
def prob4(filename="example.html"):
    """Read the specified file and load it into BeautifulSoup. Find the only
    <a> tag with a hyperlink and return its text.
    """
    #Read the html file we saved
    with open(filename, 'r') as infile:
        soup = BeautifulSoup(infile, "html.parser")
    #Find all the <a> tags
    all_a_tags = soup.find_all(name="a")
    #Give warning message if too many <a> tags
    if len(all_a_tags) > 1:
        print("There is more than one <a> tag!")
    #Return the string from the only <a> tag
    else:
        return all_a_tags[0].get_text()


# Problem 5
def prob5(filename="san_diego_weather.html"):
    """Read the specified file and load it into BeautifulSoup. Return a list
    of the following tags:

    1. The tag containing the date 'Thursday, January 1, 2015'.
    2. The tags which contain the links 'Previous Day' and 'Next Day'.
    3. The tag which contains the number associated with the Actual Max
        Temperature.

    Returns:
        (list) A list of bs4.element.Tag objects (NOT text).
    """
    #Read html into a Beautiful Soup object
    with open(filename, 'r') as infile:
        weather_html = BeautifulSoup(infile, 'html.parser')
    #Initialize list of tags to be returned
    tags = []
    #Append tag for 1.
    tags.append(weather_html.find(string="Thursday, January 1, 2015").parent)
    #Use re.compile to find the appropriate links for 2.
    tags.append(weather_html.find(string=re.compile("Previous Day")).parent)
    tags.append(weather_html.find(string=re.compile("Next Day")).parent)
    tags.append(weather_html.find(string=re.compile("^59$")).parent)
    return tags



# Problem 6
def prob6(filename="large_banks_index.html"):
    """Read the specified file and load it into BeautifulSoup. Return a list
    of the tags containing the links to bank data from September 30, 2003 to
    December 31, 2014, where the dates are in reverse chronological order.

    Returns:
        (list): A list of bs4.element.Tag objects (NOT text).
    """
    #Open the file and create a Beautiful Soup object
    with open(filename, 'r') as infile:
        soup = BeautifulSoup(infile, "html.parser")
    #Return the list of tags with links to data from September 30, 2003 - Dec 31,2014
    #using re.compile in the search function
    return [tag.parent for tag in soup.find_all(href=True, string=re.compile(r"^[A-Z][a-z]+ 3[01], 20(0[3-9]|1[0-4])$"))]


# Problem 7
def prob7(filename="large_banks_data.html"):
    """Read the specified file and load it into BeautifulSoup. Create a single
    figure with two subplots:

    1. A sorted bar chart of the seven banks with the most domestic branches.
    2. A sorted bar chart of the seven banks with the most foreign branches.

    In the case of a tie, sort the banks alphabetically by name.
    """
    #Read in the file and create a Beautiful soup object
    with open(filename, 'r') as infile:
        soup = BeautifulSoup(infile, "html.parser")

    #Grab the rows of the table, minus the header
    rows = soup.find_all(attrs={"valign":"TOP"})[1:]
    #Initialize dictionaries for banks and their number of branches
    domestic_branches = dict()
    foreign_branches = dict()
    #Assign number of branches to each bank using html parsing
    for row in rows:
        bank_name = row.td.get_text()
        domestic_branches_tag = row.td.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling.next_sibling
        #We have a few formatting obstacles in the data:
        #   1. Missing observations have a '.' instead of a number
        #   2. Numbers greater than 999 are written with a comma such as 1,484
        try:
            num_domestic_branches = int(domestic_branches_tag.get_text().replace(",",""))
            num_foreign_branches = int(domestic_branches_tag.next_sibling.next_sibling.get_text().replace(",",""))
            #For some reason there are repeats in the data so we only want to keep the
            #   largest value of domestic or foreign branches listed for any one
            #   bank.
            #Domestic branches: keep larger if duplicate
            try:
                if domestic_branches[bank_name] < num_domestic_branches:
                    domestic_branches[bank_name] = num_domestic_branches
                    #print("Updated the number of domestic branches for: " + bank_name)
                #If the current value stored in the dictionary is larger,
                #   don't update the value.
                else:
                    pass
                    #print("Prevented an update of number of domestic branches for: " + bank_name)
            #If there is no value for that bank, write in the value
            except Exception as e:
                domestic_branches[bank_name] = num_domestic_branches
            #Foreign branches: keep larger if duplicate
            try:
                if foreign_branches[bank_name] < num_foreign_branches:
                    foreign_branches[bank_name] = num_foreign_branches
                    #print("Updated the number of foreign branches for: " + bank_name)
                #If the current value stored in the dictionary is larger,
                #   don't update the value.
                else:
                    pass
                    #print("Prevented an update of number of foreign branches for: " + bank_name)
            #If there is no value for that bank, write in the new value
            except Exception as e:
                foreign_branches[bank_name] = num_foreign_branches
        #Print error if no data for that bank
        except Exception as e:
            pass
            #print("Sorry, there's been a problem for bank: " + bank_name)
            #print("It seems we don't have data on that bank.")
    #Find top domestic and foreign banks
    top_domestic = nlargest(7, domestic_branches, key=domestic_branches.get)
    domestic_freq = [domestic_branches[name] for name in top_domestic]
    top_foreign = nlargest(7, foreign_branches, key=foreign_branches.get)
    foreign_freq = [foreign_branches[name] for name in top_foreign]
    #Plot the results
    #   Domestic
    plt.figure(figsize=(10,8))
    plt.subplot(211)
    plt.title("Top Seven Banks")
    plt.barh(top_domestic, domestic_freq)
    plt.xlabel("Number of Domestic Branches")
    plt.ylabel("Bank")
    #   Foreign
    plt.subplot(212)
    plt.title("Top Seven Banks")
    plt.barh(top_foreign, foreign_freq)
    plt.xlabel("Number of Foreign Branches")
    plt.ylabel("Bank")
    plt.tight_layout(1)

    plt.show()

```
